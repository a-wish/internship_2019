```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import scipy.ndimage
```

```python
path = "C:/Users/Abhishek Tiwari/Desktop/Disparity/Test"
```

```python
imgL = cv2.imread(path + '/' + '001_L.png',0)
imgR = cv2.imread(path + '/' + '001_R.png',0)
```

```python
cv2.imshow("x",imgL)    
cv2.waitKey(0)
```

```python
import time
start = time.time()

# using OpenCV's built in library function to create a disparity map
stereo = cv2.StereoBM_create(numDisparities=48, blockSize=15)
disparity = stereo.compute(imgL,imgR)
plt.imshow(disparity,'gray')
plt.show()
end = time.time()

display("time taken to compute disparity map = " + str(end - start) +" seconds." )
```
